```python
def lag_rolling_diff(df,col):
    for i in [1,3,5,7]:
        df['lag_'+col+'_'+str(i)] = df[col].transform(lambda x: x.shift(i))

    for j in [2,3,5,7]:    
        df['rolling_mean_'+col+'_'+str(j)] = df[col].transform(lambda x: x.rolling(j).mean())

        df['rolling_std_'+col+'_'+str(j)]  = df[col].transform(lambda x: x.rolling(j).std())
        
    for k in [1,2,3]:
        df['diff_'+col+'_'+str(k)] = df[col].transform(lambda x: x.diff(k))
        df['diff_'+col+'_'+str(k)+'_ratio'] = df['diff_'+col+'_'+str(k)] / df[col]
    
    return df

lag_rolling_diff(df,'A厂')
```

```python
for k in [1,2,3]:
    df['diff_'+'col_'+str(k)] = df['A厂'].transform(lambda x: x.diff(k))
    df['diff_'+'col_'+str(k)+'_ratio'] = df['diff_'+'col_'+str(k)] / df['A厂']
```
