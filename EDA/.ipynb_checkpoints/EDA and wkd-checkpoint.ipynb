```python
import pandas as pd

df = pd.read_csv('./data/train.csv')
df.columns = ['date','A','B']
df1 = pd.read_csv('./data/wkd_v1.csv')
df1.columns = ['date','type']
df1.date = pd.to_datetime(df1.date)
df.date = pd.to_datetime(df.date)
df = df.merge(df1,on='date',how='left')
```

```python
df
```

```python
# 
def get_inner_date_feature_eng(data):
    data['date'] = pd.to_datetime(data['date'])
    data["month"] = data['date'].dt.month
    data["year"] = data['date'].dt.year

    data["day"] = data['date'].dt.day
    data["dayofw"] = data['date'].dt.dayofweek
    data["dayofy"] = data['date'].dt.dayofyear
    data["week"] = data['date'].dt.week
    data["quarter"] = data['date'].dt.quarter

    return data
```

```python
df
```

```python
k = df
k1 = k[['date','A','type']]
k2 = k[['date','B','type']]
```

```python
get_inner_date_feature_eng(k1)
```

```python
k = k1.copy()
k = get_inner_date_feature_eng(k)
train1 = k[['type',  'month',
       'year', 'day', 'dayofw', 'dayofy', 'week', 'quarter']][:-30]
train1_y = k['A'][:-30]

valid1 = k[['type',  'month',
       'year', 'day', 'dayofw', 'dayofy', 'week', 'quarter']][-30:]
valid1_y = k['A'][-30:]
```

```python
test_v1 = pd.read_csv('data/test.csv')
test_v1.columns = ['date','A','B']
test_v1.date = pd.to_datetime(test_v1.date)
test_v1 = test_v1.merge(df1,on='date',how='left')
```

```python
test_v1
```

```python
test_v1_A = test_v1[['date','A','type']]
test_v1_A = get_inner_date_feature_eng(test_v1_A)
test_v1_A
```

```python
test = test_v1_A[['type',   'month',
       'year', 'day', 'dayofw', 'dayofy', 'week', 'quarter']]


cat_cols = ['type']
train1[cat_cols] = train1[cat_cols].astype('category')
valid1[cat_cols] = valid1[cat_cols].astype('category')
test[cat_cols] = test[cat_cols].astype('category')
```

```python
import pandas as pd
from tqdm import tqdm
import warnings
import lightgbm as lgb
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import re

warnings.filterwarnings('ignore')
```

```python
params = {'learning_rate': 0.1, 
        'boosting_type': 'gbdt', 
        'objective': 'regression_l1',
        'metric': 'poisson',
        'seed': 2019, 
        'verbosity': -1, 
       }


train_set = lgb.Dataset(train1, train1_y)
val_set = lgb.Dataset(valid1, valid1_y)

model = lgb.train(params, train_set, num_boost_round=2000,
                  valid_sets=(train_set, val_set), early_stopping_rounds=50,
                  verbose_eval=50,
                 categorical_feature=cat_cols
                 )
oof_train = model.predict(valid1)
test_predict = model.predict(test)
```

```python
test_predict
```

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / (y_true))) 
test_v1_A['A'] = test_predict
mape(valid1_y, oof_train)
```

```python
train1 = k[['type',  'month',
       'year', 'day', 'dayofw', 'dayofy', 'week', 'quarter']][:-30]
train1_y = k['B'][:-30]

valid1 = k[['type',  'month',
       'year', 'day', 'dayofw', 'dayofy', 'week', 'quarter']][-30:]
valid1_y = k['B'][-30:]
```

```python
cat_cols = ['type', ]
train1[cat_cols] = train1[cat_cols].astype('category')
valid1[cat_cols] = valid1[cat_cols].astype('category')


test[cat_cols] = test[cat_cols].astype('category')

params = {'learning_rate': 0.1, 
        'boosting_type': 'gbdt', 
        'objective': 'regression_l1',
        'metric': 'poisson',
        'seed': 2019, 
        'verbosity': -1, 
       }


train_set = lgb.Dataset(train1, train1_y)
val_set = lgb.Dataset(valid1, valid1_y)

model = lgb.train(params, train_set, num_boost_round=5000,
                  valid_sets=(train_set, val_set), early_stopping_rounds=200,
                  verbose_eval=50,
                 categorical_feature=cat_cols
                 )
oof_train = model.predict(valid1)
test_predict = model.predict(test)
```
